### Plotting the above plots in a single plot using subplots

```python
cols  = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
title = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig, ax = plt.subplots(figsize=(12,30))
plt.xticks([])
plt.yticks([])

for i in range(8):
    ax = fig.add_subplot(8,1,i+1)
    ax = recent_grads[cols[i]].hist(bins=10)
    ax.set_title(title[i])
    ax.set_ylabel("Frequency")
```

### Comparison on a few major categories of men and women

```python
r = recent_grads.pivot_table(index = 'Major_category', values = ['Men', 'Women'], aggfunc = np.sum)
r
```

```python
r.plot(kind= 'bar')
plt.ylabel("Frequency")
```

#### Analytics - 
    - We observe that there is a more or less equal distribution of major categories in men and women.
    - For instance :
        - In the field of Business related majors men are predominant.
        - In the field of Health related majors women are predominant and so on.

## Extra Visualisations

### Comparing the number of men and women in each category of majors

```python
# Plotting last 10 Majors
Share.tail(10).plot.barh(x='Major')
```

#### Analytics -

## Exploring the distribution of median salaries with unemployment rate

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.boxplot(recent_grads["Unemployment_rate"].values)
plt.xlabel("Unemployment_rate")
ax.set_ylim(-0.1,0.2)

plt.show()
```

```python
num_cols = ["Unemployment_rate", "Median"]

fig, ax = plt.subplots(figsize=(8,8))
fig
ax.boxplot(recent_grads[num_cols].values)
ax.set_xticklabels(num_cols, rotation=90)

plt.show()
```

#### Analytics - 
    - There are a couple of values in the Median column that are beyond the boxplot's whiskers. These values are outliers / extreme values. These values might affect the overall observation of the dataset due to their high abnormal value.
    
    - The bottom 25% of the Median values ranges from 20000 to around 30000.
    - The top 25% of the Median values ranges from approx. 42000 to approx. 60000.
